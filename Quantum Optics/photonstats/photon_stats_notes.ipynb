# Photon Statistics Lab Notes:
In the photon statistics station of this course, our goal is to use single photon detectors to explore the underlying statistics of detection events for various kinds of light sources. This will allow us to characterize (and differentiate) different light sources by analyzing their count statistics and multi-photon detection correlations. We will characterize our detector and photon count signal (signal rise time and detector dead time) and subsequently will investigate the count statistics for dark counts and coherent light source including the poissonian error model that underlies (and is often assumed) in their distributions. We will also investigate and characterize multi-photon event correlations for various light sources and characterize our detector's jitter which limits the temporal resolution of our detector.
