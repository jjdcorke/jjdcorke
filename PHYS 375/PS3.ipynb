```python
import numpy as np
import scipy.integrate as ode
import scipy as sp
import matplotlib.pyplot as plt
import scipy.constants as const


#class to solve generic ODE
class Ode:
    def __init__(self, f, y0, h, ti, tf):

        self.y0 = y0
        self.ts = np.arange(ti,tf,h)

    #method to solve an IVP for the ODE that is passed
    def solve(self):
        return ode.odeint(self.f,self.y0, self.ts)

    
    


#class to create an embden equation with a specified n
class Embden(Ode):
    def __init__(self, n, y0 = [1, 0], h = 0.0001, ti = 0.0001, tf = 20):
        self.n = n
        self.h = h
        #initialize superclass
        Ode.__init__(self, self.f, y0, h, ti, tf)
        
        #store solution values
        solved = self.solve()
        self.thetas = solved[:,0]
        self.thetadots = solved[:,1]

    #ODE
    def f(self, y, t):
        theta = y[0]
        thetadot = y[1]
        return np.array([thetadot, -2/t*thetadot-theta**self.n])

   
    #plotter helper
    def plottheta(self):
        plt.plot(self.ts, self.thetas, label = str(self.n))


#our stack of embends
embdn0 = Embden(0)
embdn1 = Embden(1)
embdn2 = Embden(2)
embdn3 = Embden(3)
embdn4 = Embden(4)
embdn5 = Embden(5)

#list to make plotting easy
embdn = [embdn0,embdn1,embdn2, embdn3, embdn4, embdn5]

#more plotting
plt.figure(figsize=(10,10))
for i in embdn:
    i.plottheta()
plt.legend(loc = 'upper right')
plt.axhline(0).set_color('black')
plt.ylim((0,1))
plt.ylabel("Dimensionless Density").set(fontsize = 18)
plt.xlabel("x").set(fontsize = 18)
plt.show()
```

For this question, we initialize an object from our class with n = 3. We can then use root finding strategies to find where our value for theta is equal to 0 (i.e. where our value for r is Rstar and x is x_surf). Using our mapping from r to x we can then compute a numerical value for alpha using $\alpha = \frac{R_\text{star}}{x_\text{surf}}$. 
The easiest where our solution crosses is to apply the absolute value function and finding where we have a minima, $\bar{x}$. 
Then, we know that theta changes signs between this value of $\bar{x}$ and $\bar{x}\pm h$ depending if xbar is positive or negative. 
Once we know the sign (in this case positive),we can assume it lies between xbar and xbar+h and take an average of these and state that our root is $x_0 \approx \bar{x}+\frac{1}{2}h \pm \frac{h}{2}$. 
Then we have our mapping bvetween x and r giving us $\alpha = \frac{r}{x}$ which in our case we substitute appropriate values of r and x namely, our root and Rsun. 

```python

#find index where theta is closest to 0
argroot = np.argmin(np.abs(embdn3.thetas))

#sign check to make sure we take the right average
if embdn3.thetas[argroot] > 0:
    sign = "positive"

else:
    sign = "negative"

print("The sign of xbar is " + sign)

#find the value of our root
root = embdn3.ts[argroot] + embdn3.h/2

print("Theta is equal to 0 at x: ", root, " +-", embdn3.h/2)


Rsun = 6.95700e8 #from google

#value for alpha
alpha = Rsun/root

print("The numerically computed value of alpha we get is: ", alpha)

#needed for rhoc
dthetaroot = (embdn3.thetadots[argroot]+embdn3.thetadots[argroot+1])/2

print("Dthetadx at our root is: ", dthetaroot)
Msun = 1.98855e30 #from google

#value for rhoc
rhoc = -Msun/(4*np.pi*(alpha**3)*(root**2)*dthetaroot)

print("rhoc is then: ", rhoc)

#value for K
K = (alpha**2*np.pi*const.G)/(rhoc**(-2/3))
print("The constant K is: ", K)
```
